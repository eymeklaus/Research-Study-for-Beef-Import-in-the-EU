```python
consumption = pd.read_csv(r"C:\Users\eymem\Documents\GitHub\Project-Beef\consumption_new.csv")
price_year = pd.read_csv(r"C:\Users\eymem\Documents\GitHub\Project-Beef\price_year.csv")
```

```python
comparison= consumption.merge(price_year, how = "outer")
```

```python
comparison.head()
```

```python
comparison.shape
```

```python
comparison.isna().sum()
```

```python
comparison["Price €/kg"] = (comparison["Price €/100kg"]/ 100).round(2)
```

```python
comparison.head()
```

```python
pivot_df = comparison.pivot_table(index='Country', columns='Year', values=["Price €/kg"])
pivot_df
```

```python
comparison["Per Capita Beef Consump."] = comparison["Per Capita Beef Consump."].round(2)
```

```python
pivot_df = comparison.pivot_table(index='Country', columns='Year', values=["Per Capita Beef Consump."])
pivot_df
```

```python
comparison["Country"].value_counts()
```

```python
comparison["Year"].value_counts()
```

```python
comparison.describe()
```

```python
comparison["Total Expenditure"] = comparison["Price €/kg"] * comparison["Per Capita Beef Consump."]
```

```python
comparison.head()
```

```python
#Total_Expenditure mean from 2013 -2022

total_expenditure = (comparison[(comparison['Year'] >= 2013) & (comparison['Year'] <= 2022)].groupby('Country')['Total Expenditure'].mean().reset_index().round(2))

print(total_expenditure)
```

```python
plt.figure(figsize=(15,7))

plt.bar(
    total_expenditure['Country'],
    total_expenditure['Total Expenditure']
)

plt.xticks(rotation=45)
plt.title("Average Beef Expenditure from 2013 - 2022")
plt.xlabel("Country")
plt.ylabel("Expenditure")
plt.tight_layout()
plt.show()
```

```python
def top_ten(total_expenditure):
    return total_expenditure.nlargest(5, "Total Expenditure")
```

```python
top_ten(total_expenditure)
```

```python
countries = ["Luxembourg", "Sweden", "Denmark", "France", "Ireland"]


pivot_price = (comparison[(comparison['Country'].isin(countries)) & (comparison['Year'].between(2013, 2022))].pivot_table(index='Year', columns='Country', 
            values='Price €/kg'))

print(pivot_price)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))

for country in pivot_price.columns:
    plt.plot(pivot_price.index,
             pivot_price[country],
             marker='o',
             linewidth=2,
             label=country)

plt.title('Price Beef by Country', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Price €/kg')
plt.legend(
    title='Country',            # optional title for legend
    fontsize=10,                # smaller text
    title_fontsize=12,          # title size
    bbox_to_anchor=(1.05, 1),   # move legend to the right outside plot
    loc='upper left')            # anchor position

plt.tight_layout()
plt.show()
```

```python
pivot_consum = (comparison[comparison['Country'].isin(countries) & (comparison['Year'].between(2013, 2022))].pivot_table(index='Year',columns='Country',
                values="Per Capita Beef Consump."))

print(pivot_consum)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))

for country in pivot_consum.columns:
    plt.plot(pivot_consum.index,
             pivot_consum[country],
             marker='o',
             linewidth=2,
             label=country)

plt.title('Per Capita Consumption Beef kg/year', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Per Capita Consumption')
plt.legend(
    title='Country',            # optional title for legend
    fontsize=10,                # smaller text
    title_fontsize=12,          # title size
    bbox_to_anchor=(1.05, 1),   # move legend to the right outside plot
    loc='upper left')            # anchor position

plt.tight_layout()
plt.show()
```
