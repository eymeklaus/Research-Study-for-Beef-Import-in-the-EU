```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=comparison,
    x='Year',
    y='Price €/100kg',
    marker='o'
)

plt.title('Average Price per Year', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Average Price €/100kg')

plt.grid(True)
plt.show()
```

```python
pivot_df = comparison.pivot_table(index='Year', columns='Country', values=['Price €/100kg'])
pivot_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,12))

for country in pivot_df.columns:
    plt.plot(pivot_df.index,
             pivot_df[country],
             marker='o',
             linewidth=2,
             label=country)

plt.title('Price Behavior Over Time by Country', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Price €/100kg')
plt.legend(
    title='Country',            # optional title for legend
    fontsize=10,                # smaller text
    title_fontsize=12,          # title size
    bbox_to_anchor=(1.05, 1),   # move legend to the right outside plot
    loc='upper left'            # anchor position

plt.tight_layout()
plt.show()
```
